```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
pd.options.display.max_columns = None
```

```python
ww = pd.read_csv("Summary of Weather.csv")
ww.head()
```

```python
ww.shape
```

```python
ww.info()
```

```python
temp = pd.Series(ww.isna().sum()/ww.shape[0] > 0.8)
temp_ = temp[temp == True].index
ww.drop(temp_, axis=1, inplace=True)
```

```python
ww.head()
```

```python
ww.info()
```

```python
ww['Date'] = pd.to_datetime(ww['Date'])
```

```python
temp = ww.Precip.str.isdigit()
temp_ = temp[temp == False].index
ww.iloc[temp_]
```

```python
ww['Precip'] = ww['Precip'].replace('T', np.nan)
```

```python
ww['Precip'] = ww['Precip'].astype('float64')
```

```python
ww['Snowfall'].replace("#VALUE!", np.nan, inplace=True)
```

```python
ww["Snowfall"] = ww['Snowfall'].astype('float64')
```

```python
ww.info()
```

```python
ww[ww['PoorWeather'] == '1     1']
```

```python
ww['PoorWeather'].value_counts()
```

```python
ww.isna().sum()
```

```python
ww.drop(['PoorWeather', 'TSHDSBRSGF'], axis=1, inplace=True)
```

```python
ww.info()
```

```python
ww['PRCP'].replace('T', np.nan, inplace=True)
```

```python
ww['PRCP'] = ww['PRCP'].astype('float64')
```

```python
ww['SNF'].replace('T', np.nan, inplace=True)
```

```python
ww['SNF'] = ww['SNF'].astype('float64')
```

```python
ww.info()
```

```python
ww = ww.apply(lambda x:x.fillna(x.median(), axis=0))
```

```python
ww.isna().sum()
```

```python
sns.pairplot(ww)
```

```python
ww.drop(['Date', 'DA'], axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = ww.loc[:, ww.columns != 'MaxTemp']
y = ww[['MaxTemp']]

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression_model = LinearRegression()
regression_model.fit(X_train, y_train)
```

```python
print("Train score", regression_model.score(X_train, y_train))
print("Test score", regression_model.score(X_test, y_test))
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2, interaction_only=True)
```

```python
X_train_ = poly.fit_transform(X_train)
X_test_ = poly.fit_transform(X_test)
```

```python
poly_clf = LinearRegression()
poly_clf.fit(X_train_, y_train)
```

```python
print("Train score", poly_clf.score(X_train_, y_train))
print("Test score", poly_clf.score(X_test_, y_test))
```

```python
X_train_.shape
```

```python
from sklearn.linear_model import Ridge, Lasso
```

```python
ridge = Ridge(alpha=0.3)
ridge.fit(X_train_, y_train)
print("Ridge score:", ridge.score(X_train_, y_train))
print("Ridge score:", ridge.score(X_test_, y_test))
```

```python
lasso = Lasso(alpha=0.1)
lasso.fit(X_train_, y_train)
print("Lasso score:", lasso.score(X_train_, y_train))
print("Lasso score:", lasso.score(X_test_, y_test))
```
